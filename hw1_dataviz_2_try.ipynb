```python
# import libraries
import sys, csv

import altair as alt
import pandas as pd
```

```python
# read excel
df = pd.read_excel('GLB.Ts_dSST.xlsx')
df.head()
```

```python
df.head()
```

```python
df['Land-Ocean: Global Means'][0].split(",")
```

```python
data = []
for i in range(1, len(df)):
    year = df['Land-Ocean: Global Means'][i].split(",")[0]
    temperatures = df['Land-Ocean: Global Means'][i].split(",")[1:13]
    e = 0
    for t in temperatures:
        e += float(t)
    e = e/12
    data.append((year, e))
```

```python
data
```

```python
df1 = pd.DataFrame(data, columns =['Year', 'Temperature'])
```

```python
df1.head()
```

```python
years_to_show = ['1904','1944','1998','2014','2015','2016','2017']
bar = alt.Chart(df1).mark_circle(color='orange', size=80).encode(
    x=alt.X('Year:Q',scale=alt.Scale(domain=[1880,2020]), axis=alt.Axis(title=None,values=['1880','1890','1900','1910','1920','1930','1940','1950','1960','1970','1980','1990','2000','2010'])),
    y=alt.Y('Temperature:Q', axis=alt.Axis(title=None)),
).properties(title = alt.TitleParams(["Annual Global Surface Temperature","Relative to Late Century Average"], anchor='start', ))
rule = alt.Chart(df1).mark_rule(color='black').encode(
    y='mean(Temperature)'
)
text = alt.Chart(df1).mark_text(dx=20, dy=0).encode(
    x=alt.X('Year:Q', aggregate={'argmax': 'Temperature'}),
    y=alt.Y('max(Temperature):Q'),
    text=alt.condition(alt.datum.Year >= '2014', 'Year:Q', alt.value(''))
)
text1 = alt.Chart(df1).mark_text(dx=20, dy=0).encode(
    x=alt.X('Year:Q', aggregate={'argmax': 'Temperature'}),
    y=alt.Y('max(Temperature):Q'),
    text=alt.condition(alt.datum.Year == '1944', 'Year:Q', alt.value(''))
)
text2 = alt.Chart(df1).mark_text(dx=-20, dy=0).encode(
    x=alt.X('Year:Q', aggregate={'argmax': 'Temperature'}),
    y=alt.Y('max(Temperature):Q'),
    text=alt.condition(alt.datum.Year == '1998', 'Year:Q', alt.value(''))
)
text3 = alt.Chart(df1).mark_text(dx=-20, dy=0).encode(
    x=alt.X('Year:Q', aggregate={'argmax': 'Temperature'}),
    y=alt.Y('max(Temperature):Q'),
    text=alt.condition(alt.datum.Year == '1904', 'Year:Q', alt.value(''))
)
text4 = alt.Chart(df1).mark_text(dx=425,dy=20).encode(
    y='mean(Temperature)',
    text=alt.value('COLDER')
)
text5 = alt.Chart(df1).mark_text(dx=395,dy=-40).encode(
    y='mean(Temperature)',
    text=alt.value('HOTTER THAN THE')
)
text6 = alt.Chart(df1).mark_text(dx=390,dy=-20).encode(
    y='mean(Temperature)',
    text=alt.value('1880-1899 AVERAGE')
)
text7 = alt.Chart(df1).mark_text(dx=-450,dy=300, color='gray').encode(
    text=alt.value('source: NASA')
)

(bar+rule+text+text1+text2+text3+text4+text5+text6+text7).configure_view(width = 900, height = 500)
```
